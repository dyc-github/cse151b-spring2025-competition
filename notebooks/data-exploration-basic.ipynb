## Start

```python
import zarr
import xarray as xr
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
# Load the data (update the path to the correct location)
data_path = "./processed_data_cse151b_v2_corrupted_ssp245/processed_data_cse151b_v2_corrupted_ssp245.zarr"
data = xr.open_zarr(data_path)
data
```

## Train Test Size


```python
train_sizes = data.sel(ssp=["ssp126", "ssp370", "ssp585"]).sizes
train_size= train_sizes['ssp'] * train_sizes['time'] * train_sizes['x'] * train_sizes['y'] * train_sizes['member_id']
display(train_size)

test_sizes = data.sel(ssp=["ssp245"]).sizes
test_size= test_sizes['ssp'] * test_sizes['time'] * test_sizes['x'] * test_sizes['y'] * test_sizes['member_id']
display(test_size)
```

## Distribution of Data

### Output data

```python
for var in ['tas', 'pr']:
  print(var)
  data_var = data.sel(ssp=["ssp126", "ssp370", "ssp585"])[var].to_numpy().flatten()
  display(pd.DataFrame(data_var).describe())
```

```python
bins = np.arange(200, 320, 5)
data.sel(ssp=["ssp126", "ssp370", "ssp585"]).tas.plot.hist(bins = bins)
plt.ylabel("Count")
plt.show()
```

```python
bins = np.arange(0, 70, 2)
data.sel(ssp=["ssp126", "ssp370", "ssp585"]).pr.plot.hist(bins = bins)
plt.ylabel("Count")
plt.show()
```

```python
# BC
# CH4
# CO2
# SO2
# rsdt
for var in ['BC', 'CH4', 'CO2', 'SO2', 'rsdt']:
  print(var)
  data_var = data.sel(ssp=["ssp126", "ssp370", "ssp585"])[var].to_numpy().flatten()
  display(pd.DataFrame(data_var).describe())


  # mean = data.sel(ssp=["ssp126", "ssp370", "ssp585"])[var].mean(skipna=True).to_numpy()
  # print('mean:', mean.item())
  # std = data.sel(ssp=["ssp126", "ssp370", "ssp585"])[var].std(skipna=True).to_numpy()
  # print('standard deviation:', std.item())
```

```python
#BC: Heavily right skewed distribution
bins = np.arange(0, 5e-11, 1.3e-12)
data.sel(ssp=["ssp126", "ssp370", "ssp585"]).BC.plot.hist(bins = bins)
plt.ylabel("Count")
plt.show()
```

## Plot a few snapshots of temperature and precipitation (first, middle, and last)

```python
data_subset = data.isel(time=[0, data.time.size // 2, -1]).sel(ssp="ssp370", member_id=0)
data_subset.tas.plot(col="time", x="x", y="y", robust=True)
data_subset.pr.plot(col="time", x="x", y="y", robust=True)
```

# Plot the time series of global mean temperature and precipitation

Note that we do not plot the SSP245 scenario here, as it is corrupted in the provided data.

```python
global_mean_data = data.sel(ssp=["ssp126", "ssp370", "ssp585"]).mean(dim=["x", "y", "member_id"])  # also average over all ensemble members
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
global_mean_data.tas.plot(label="Temperature", ax=ax[0], hue="ssp")
global_mean_data.pr.plot(label="Precipitation", ax=ax[1], hue="ssp")# Plot the time series of global mean temperature and precipitation
```

# Plot the (global mean) time series of climate forcings

```python
global_mean_data = data.sel(ssp=["ssp126", "ssp370", "ssp585"]).mean(dim=["longitude", "latitude", "member_id"])
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.flatten()
global_mean_data.CO2.plot(ax=axes[0], hue="ssp")
global_mean_data.CH4.plot(ax=axes[1], hue="ssp")
global_mean_data.SO2.plot(ax=axes[2], hue="ssp")
_ = global_mean_data.BC.plot(ax=axes[3], hue="ssp")
```
